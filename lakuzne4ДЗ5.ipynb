## Задание 2.2
Какова доля совершеннолетних пассажиров (18 лет и старше)?

```python
def is_adult(Age):
    if Age>=18:
        return 1
    else:
        return 0
#g- пассажиры с указанным возрастом
g=df[df['Age'].notnull()]

answer1=g.loc[g['Age'].apply(is_adult)==1].count()['Name']/g.count()['Name']
answer2=(df.loc[df['Age'].apply(is_adult)==1].count()['Name']/df.count()['Name'])

print '%s среди пассажиров с указанным возрастом' % answer1
print '%s среди всех пассажиров' % answer2
```

## Задача 2.4
Какова доля выживших среди женщин? Среди мужчин?

```python
#доля выживших

def survival_rate(field_name,field_value):
    try:
        return df.loc[(df[field_name]==field_value)&(df.Survived==1)].Name.count()/df.loc[(df[field_name]==field_value)].Name.count()
    except:
        return 'error message'
    
print survival_rate('Sex','female')
print survival_rate('Sex','male')

# print survival_rate('Pclass',1)
# print survival_rate('Pclass',3)
```
